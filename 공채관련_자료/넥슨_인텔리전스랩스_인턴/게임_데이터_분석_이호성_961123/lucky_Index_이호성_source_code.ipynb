```python
# 3. 음이항 분포로 n회 성공하기 위한 시도 횟수를 파악합니다.
```

```python
# 성공 확률이 p인 퀘스트를 k번 성공하기 위해서 n번의 시도 수가 필요할 확률을 구한다.

nbi_list = []

for i in range(len(df4)):
   if df4['get / try'][i] != 1:
       nbi_list.append(nbinom(df4['try_count'][i]-df4['get_count'][i], 0.96).pmf(df4['try_count'][i]))
   else:
       nbi_list.append(nbinom(df4['get_count'][i],0.04).pmf(df4['try_count'][i]))
nbi_list

# 단순 확률을 구하는 것은 의미가 없고, 누적해서 얼마정도의 시도횟수가 필요한가를 구하는 것이 중요하다는 것을 파악했습니다.
```

```python
# 데이터 프레임 0번 인덱스에 대한 음이항 분포에 대해서 그려보겠습니다.
x = np.arange(df6['try_count'][0]-df6['get_count'][0])
y1 = nbinom(df6['get_count'][0],.04).pmf(x)
plt.figure(figsize=(12, 8))
plt.plot(x, y1, 'ro--')
plt.xlabel('X', fontsize=20) # get_count 번을 얻기 위해서 실패한 횟수.
plt.ylabel('P(X)', fontsize=20) # X에 대한 확률.
plt.title('Negative Binomial Distribution(n = 2, p = .04)', fontsize=25)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid()
plt.show()
```

```python
# 데이터프레임 0번 인덱스에 대해서 50%, 80%, 90%, 99%의 시도횟수를 구해보겠습니다.
```

```python
print("데이터프레임 0번 인덱스의 try_count :", df['try_count'][0], ", get_count :",df['get_count'][0])
```

```python
# 약 50%의 사람은 이 시도 횟수로 n회 이상 출현합니다.
sum = 0
index = 0
for i in y1:
    if sum >= 0.5:
        break
    sum += i
    index += 1
print("약 50% 확률로 성공할 시도 횟수는 :", index, "회 입니다.") 
```

```python
# 약 80%의 사람은 이 시도 횟수로 n회 이상 출현합니다.
sum = 0
index = 0
for i in y1:
    if sum >= 0.8:
        break
    sum += i
    index += 1
print("약 80% 확률로 성공할 시도 횟수는 :", index, "회 입니다.") 
```

```python
# 약 90%의 사람은 이 시도 횟수로 n회 이상 출현합니다.
sum = 0
index = 0
for i in y1:
    if sum >= 0.9:
        break
    sum += i
    index += 1
print("약 95% 확률로 성공할 시도 횟수는 :", index, "회 입니다.") 
```

```python
# 약 99%의 사람은 이 시도 횟수로 n회 이상 출현합니다.
sum = 0
index = 0
for i in y1:
    if sum >= 0.99:
        break
    sum += i
    index += 1
print("약 99% 확률로 성공할 시도 횟수는 :", index, "회 입니다.") 
```

```python
# 99% 확률로 성공할 시도 횟수를 기준으로 행운지수를 계산하도록 하겠습니다.
```

```python
# 1. 99% 확률로 'get_count'회 뽑을 수 있는 음이항 분포상의 값을 구함
# 2-1. try_count / get_count회 뽑기 위한 99% 음이항 분포 값으로 행운지수를 나타냄
# 2-2 try_count / get_count가 0인 경우에는 1회 뽑기 위한 시도횟수(99% 음이항분포)으로 행운지수를 나타냄
# 3. 해당 행운 지수가 담겨있는 리스트를 데이터프레임으로 만듬.
# cf. 해당 연산이 오래걸려서 진행정도를 파악하기 위해 tqdm 라이브러리를 적용하여 진행정도를 파악하였습니다.


# n회 성공하기 위한 99% 이상의 시도횟수 구하기.
index_set_99 = []
for i in notebook.tqdm(range(max(df['get_count']) + 1), desc="iterate range 100"):
    index = 1
    if i == 0:
        for j in range(200):
            y1 = nbinom(1, .04).cdf(j)
            if y1 >= 0.99:
                break
            index += 1
        index_set_99.append(index)
            
    else :
        for j in range(3000):
            y1 = nbinom(i, .04).cdf(j)
            if y1 >= 0.99:
                break
            index += 1
        index_set_99.append(index)
```

```python
# 연산 수행.
ninety_nine = []

for i in notebook.tqdm(range(len(df)), desc="iterate range 100"):
        ninety_nine.append(df['try_count'][i] / index_set_99[df['get_count'][i]])

df7 = pd.DataFrame(ninety_nine)
#     else:
#         y1 = nbinom(df6['try_count'][i],.99).pmf(0)
#         y1 = 1 - y1
#         ninety_nine.append(y1)

# print(ninety_nine)
```
