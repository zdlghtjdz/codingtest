```python
# 마을 가로 70km, 세로 30km

# 마을 지도 정보를 입력받고 원하는 별장 부지를 지도에 그리는 프로그램.

town_map = {'mount' : [] , 'lake' : [], 'villa' : []} # 마을지도의 정보.

# 좌표는 좌측 상단, 우측 하단 두개의 점을 튜플로 저장
# 산이 두개인 경우
# town_map['mount'] -> [[(0, 0), (10, 10)], [(60, 20), (70, 30)]]


# 좌표가 사각형을 형성하는 지, 호수와 산이 겹치는 지 확인해야 함.

def square_check(x1,y1,x3,y3, mount): # 좌표가 유효한지 확인하는 함수
    
    # 유효 좌표 체크
    for i in mount:
        if i > 100 or i < 0 :
            return 0
    
    #사각형 형성 체크
    if x1 >= x3 :
        return 0
    elif y1 >= y3 :
        return 0
    else:
        return 1

def print_map(m): # 맵 출력 함수
    for i in m:
        for j in i :
            print(j, end='')
        print()
    
def lake_check(m, mount): # 산 - 산, 호수 - 호수, 산 - 호수, 호수 - 산 체크
    x1, y1 = mount[0] + 1, mount[1] + 1 #x2 y2 = x3, y1
    x3, y3 = mount[2] + 1, mount[3] + 1 #x4 y4 = x1, y3
    
    for i in range(y1, y3):
        for j in range(x1, x3):
            if m[i][j] == "~" :
                return 0
    
    return 1

def mountain_check(m, villa):
    x1, y1 = villa[0] + 1, villa[1] + 1 #x2 y2 = x3, y1
    x3, y3 = villa[2] + 1, villa[3] + 1 #x4 y4 = x1, y3
    
    for i in range(y1, y3):
        for j in range(x1, x3):
            if m[i][j] == "^":
                return 0
        
m = [[" " for _ in range(72)] for _ in range(32)]

for i in range(len(m)) :
    if i == 0 or i == 31 :
        for j in range(len(m[i])):
            m[i][j] = "*"
    else:
        for j in range(len(m[i])):
            if j == 0 or j == len(m[i]) - 1:
                m[i][j] = "*"

while True:
    print("--------------------------")
    print("별장 지도 프로그램 입니다.")
    print("--------------------------")
    print("1. 산 위치 입력")
    print("2. 호수 위치 입력")
    print("3. 별장 위치 입력")
    print("4. 마을 지도 출력")
    print("0. 종료")
    print("옵션을 선택해주세요. : ")
    a = int(input())
    print("--------------------------")
    
    if a == 1 :
        print("산의 왼쪽 상단(x1, y1)과 오른쪽 하단(x3, y3) 좌표를 띄어쓰기로 입력해주세요. :") # 산 : ^
        mount = list(map(int, input().split()))
               
        x1, y1 = mount[0] + 1, mount[1] + 1 #x2 y2 = x3, y1
        x3, y3 = mount[2] + 1, mount[3] + 1#x4 y4 = x1, y3
        
        if mountain_check(m, mount) == 0:
            print("산의 면적이 겹칩니다.")
            break
        
        if square_check(x1, y1, x3, y3, mount) == 0:
            print("좌표가 명확하지 않습니다.")
            break
                    
        town_map['mount'].append([(x1, y1), (x3, y3)])
        
        for i in range(y1, y3):
            for j in range(x1, x3):
                m[i][j] = "^"
                           
            
    elif a == 2 :
        print("호수의 왼쪽 상단(x1, y1)과 오른쪽 하단(x3, y3) 좌표를 띄어쓰기로 입력해주세요. :") # 호수 : ~
        lake = list(map(int, input().split()))
        
        x1, y1 = lake[0] + 1, lake[1] + 1 #x2 y2 = x3, y1
        x3, y3 = lake[2] + 1, lake[3] + 1#x4 y4 = x1, y3
        
        if square_check(x1, y1, x3, y3, lake) == 0:
            print("좌표가 명확하지 않습니다.")
            break
            
        if lake_check(m, lake) == 0:
            print("호수의 면적이 겹칩니다.")
            break
        
        town_map['lake'].append([(x1, y1), (x3, y3)])
        
        for i in range(y1, y3):
            for j in range(x1, x3):
                m[i][j] = "~"
        
        
        
    elif a == 3 :
        print("별장의 왼쪽 상단(x1, y1)과 오른쪽 하단(x3, y3) 좌표를 띄어쓰기로 입력해주세요. :") # 별장 : o
        villa = list(map(int, input().split()))
        
        x1, y1 = villa[0] + 1, villa[1] + 1 #x2 y2 = x3, y1
        x3, y3 = villa[2] + 1, villa[3] + 1#x4 y4 = x1, y3
        
        if square_check(x1, y1, x3, y3, villa) == 0:
            print("좌표가 명확하지 않습니다.")
            break
            
        if town_map['villa'] != [] or mountain_check(m, villa) == 0:
            print("이미 별장 부지를 정했거나, 산과 면적이 겹칩니다.")
            break
        
        town_map['villa'].append([(x1, y1), (x3, y3)])
        
        for i in range(y1, y3):
            for j in range(x1, x3):
                m[i][j] = "o"
        
        
        
    elif a == 4 :
        print_map(m)
        
        
    else :
        break



```
