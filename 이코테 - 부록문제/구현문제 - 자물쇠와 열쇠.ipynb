```python
# 73점.

def solution(key, lock):
    success = 0 
    
    map_ = [[0 for _ in range(len(key) - 2 + len(lock) + len(key))] 
            for _ in range(2 * len(key) + len(lock) - 2)]
    
    key_set = []
    key_set.append(key)
    
    key_90 = [[0 for _ in range(len(key))] for _ in range(len(key)) ]
    for j in range(len(key)) :
        for k in range(len(key)) :
            key_90[j][k] = key[len(key) - k - 1][j]
        key_set.append(key_90)
    
    key_180 = [[0 for _ in range(len(key))] for _ in range(len(key)) ]
    for j in range(len(key)) :
        for k in range(len(key)) :
            key_180[j][k] = key_90[len(key) - k - 1][j]
    key_set.append(key_180)
    
    key_270 = [[0 for _ in range(len(key))] for _ in range(len(key)) ]
    for j in range(len(key)) :
        for k in range(len(key)) :
            key_270[j][k] = key_180[len(key) - k - 1][j]
    key_set.append(key_270)
        
    
    #return key_set
    
    # map 가운데 lock 초기화
    
    for i in range(len(key) - 1, len(key) + len(lock) - 1) :
        for j in range(len(key) - 1, len(key) + len(lock) - 1) :
            map_[i][j] = lock[i- len(key) + 1][j - len(key) + 1]
    
    # key, key_90, key_180, key_270를 합하기.
    
    for k in range(len(key) + len(lock) - 2) :
        for l in range(len(key) + len(lock) - 2) :
            map_copy = map_.copy()
            for n in range(k, k + len(key)) :
                for m in range(l, l + len(key)) :
                    for key_ in key_set:
                        map_copy[n][m] = key_[n-k][m-l]   
                     # 검사
                    for i in range(len(key) -1, len(key)-1 + len(lock)) :
                        if success == 1 :
                            break
                        for j in range(len(key) -1, len(key)-1 + len(lock)) :
                             
                            if map_copy[i][j] == 0 or map_copy[i][j] == 2 :
                                success = 0
                                continue
                            elif map_copy[i][j] == 1 :
                                success = 1
    
    if success == 1 :
        return True
    else :
        return False
```
